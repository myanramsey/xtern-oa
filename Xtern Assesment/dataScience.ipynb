# Data Sceince Xtern Assesment 

```python
# import libraries 
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler #Library used to normalize values for analysis 
import seaborn as sns
import matplotlib.pyplot as plt
```

## Questions to answer 
 - Which users are most likely to purchase virtual merchandise? 
 - How do user activities during race events correlate with their merchandise purchases and sponsorship interaction? 

## Read CSV


```python
df = pd.read_csv("VeloCityX Fan Engagement Data.csv")
df.head()
```

## Clean Data


- Remove any duplicate entries or users with missing data.
- Normalize or scale numerical features to ensure they're on the same scale.
- Create derived features that might be more indicative of user engagement.

### Drop Duplicates and Null Values

```python
df.drop_duplicates(keep="first", inplace= True)
# No Null Values Found
```

### Normalize Values

```python
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Define the numerical features to normalize
normalized_features = ['Fan Challenges Completed', 'Predictive Accuracy (%)', 
                       'Virtual Merchandise Purchases', 'Sponsorship Interactions (Ad Clicks)', 
                       'Time on Live 360 (mins)', 'Real-Time Chat Activity (Messages Sent)']

# Apply Min-Max scaling to the selected features
df[normalized_features] = scaler.fit_transform(df[normalized_features])

```

```python
df.head()
```

# Visualize 

```python
sns.histplot(df['Virtual Merchandise Purchases'], bins=10, kde=True)
plt.title('Distribution of Virtual Merchandise Purchases')
plt.xlabel('Number of Purchases')
plt.ylabel('Frequency')
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()

# Plot the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of User Engagement Metrics')
plt.show()
```

- Virtual Merchandise Purchases have the strongest positive correlation (0.16) with Fan Challenges Completed.
- There's a very weak positive correlation (0.07) with Sponsorship Interactions (Ad Clicks).
- Surprisingly, there's a slight negative correlation (-0.04) with Real-Time Chat Activity.

```python
# Pairplot to visualize relationships between features
sns.pairplot(df, vars=normalized_features)
plt.show()
```

# Conclusion: 


## Findings
- Users who completed more fan challenges were more likely to purchase virtual merchanidise 
- Time spent on Live 360 did not have strong correlation on items bought. 
- Predictive accuracy did not show signaficant impact on purchase. 

However it is important to note that there was no signiicantly strong correlation found, just some correlation. Meaning more data needs to be collected to find and observe strong substainial relationships.
